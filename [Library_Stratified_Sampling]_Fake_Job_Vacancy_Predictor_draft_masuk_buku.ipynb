```python
# Fungsi untuk menghitung mean dari TF-IDF tiap sel
def calculate_tfidf_mean(column):
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(column).toarray()
    tfidf_mean = tfidf_matrix.mean(axis=1)
    return tfidf_mean

# Dataframe baru untuk menyimpan hasil
tfidf_mean_df = pd.DataFrame()
# Menghitung mean TF-IDF untuk setiap kolom dan menyimpan hasilnya ke dataframe baru
for i in text_columns:
    tfidf_mean_df[i + '_numeric'] = calculate_tfidf_mean(data[i])

# # Menghitung matriks korelasi
correlation_matrix = tfidf_mean_df.corr()
correlation_matrix['function_numeric']
# correlation_matrix.abs().mean()
# print(correlation_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import log_loss

# Hitung probabilitas prediksi
y_pred_proba = nb_classifier.predict_proba(Xcount_test_transformed_ngr)

# Parameter
sample_sizes = np.linspace(10, len(y_test), num=50, dtype=int)
n_iterations = 30  # jumlah pengulangan random sampling per sample size
np.random.seed(42)  # supaya hasilnya konsisten

mean_losses = []
std_losses = []

for size in sample_sizes:
    losses = []
    for _ in range(n_iterations):
        random_indices = np.random.choice(len(y_test), size=size, replace=False)
        y_sample = y_test.iloc[random_indices] if hasattr(y_test, 'iloc') else y_test[random_indices]
        y_pred_sample = y_pred_proba[random_indices]

        loss = log_loss(y_sample, y_pred_sample)
        losses.append(loss)

    mean_losses.append(np.mean(losses))
    std_losses.append(np.std(losses))

# Rata-rata log loss keseluruhan
avg_loss = log_loss(y_test, y_pred_proba)

# Plot log loss dengan error band
plt.figure(figsize=(8,5))
plt.plot(sample_sizes, mean_losses, marker='o', linestyle='-', label="Mean Log Loss")
plt.fill_between(sample_sizes,
                 np.array(mean_losses) - np.array(std_losses),
                 np.array(mean_losses) + np.array(std_losses),
                 color='orange', alpha=0.3, label="±1 Std Dev")

plt.axhline(avg_loss, color='red', linestyle=':', label=f"Avg Log Loss = {avg_loss:.4f}")
plt.xlabel("Number of Samples")
plt.ylabel("Log Loss")
plt.title(f"Log Loss vs Number of Samples\n(Naive Bayes, {n_iterations}x Random Sampling)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, recall_score
import numpy as np

#Stratified K-Fold to Randomize the Data
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

#Predictions
y_train_pred = nb_classifier.predict(Xcount_train_transformed_ngr)
y_test_pred = nb_classifier.predict(Xcount_test_transformed_ngr)

#Evaluate accuracy
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Testing Accuracy: {test_acc:.4f}")

#Manually perform cross-validation with recall scoring
cv_scores = []

for train_idx, val_idx in cv.split(Xcount_train_transformed_ngr, y_train):
    X_fold_train, X_fold_val = Xcount_train_transformed_ngr[train_idx], Xcount_train_transformed_ngr[val_idx]
    y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

    #Train the classifier on this fold
    nb_classifier.fit(X_fold_train, y_fold_train)

    #Predict on the validation fold
    y_fold_pred = nb_classifier.predict(X_fold_val)

    #Calculate recall
    recall = recall_score(y_fold_val, y_fold_pred, average="macro", zero_division=1)
    cv_scores.append(recall)

    # # Debugging: Check unique labels predicted
    # print(f"Fold {len(cv_scores)} - Unique Predictions: {np.unique(y_fold_pred)}")

cv_scores = np.array(cv_scores)

#Calculate mean and standard deviation
mean_cv_score = np.mean(cv_scores)
std_cv_score = np.std(cv_scores)

#Threshold for overfitting detection
threshold = mean_cv_score - 2 * std_cv_score #Normal assumption

print(f"Cross-validation Recall Scores: {cv_scores}")
print(f"Cross-validation Mean Recall: {mean_cv_score:.4f}")
print(f"Cross-validation Recall Std Dev: {std_cv_score:.4f}")
print(f"Threshold for Overfitting Detection: {threshold:.4f}")

#Check for overfitting
if test_acc < threshold:
    print("Potential overfitting detected!")
else:
    print("Model generalizes well.")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, recall_score
import numpy as np

#Stratified K-Fold to Randomize the Data
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

#Predictions
y_train_pred = LR_classifier.predict(Xcount_train_transformed_ngr)
y_test_pred = LR_classifier.predict(Xcount_test_transformed_ngr)

#Evaluate accuracy
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Testing Accuracy: {test_acc:.4f}")

#Manually perform cross-validation with recall scoring
cv_scores = []

for train_idx, val_idx in cv.split(Xcount_train_transformed_ngr, y_train):
    X_fold_train, X_fold_val = Xcount_train_transformed_ngr[train_idx], Xcount_train_transformed_ngr[val_idx]
    y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

    #Train the classifier on this fold
    LR_classifier.fit(X_fold_train, y_fold_train)

    #Predict on the validation fold
    y_fold_pred = LR_classifier.predict(X_fold_val)

    #Calculate recall
    recall = recall_score(y_fold_val, y_fold_pred, average="macro", zero_division=1)
    cv_scores.append(recall)

    # # Debugging: Check unique labels predicted
    # print(f"Fold {len(cv_scores)} - Unique Predictions: {np.unique(y_fold_pred)}")

cv_scores = np.array(cv_scores)

#Calculate mean and standard deviation
mean_cv_score = np.mean(cv_scores)
std_cv_score = np.std(cv_scores)

#Threshold for overfitting detection
threshold = mean_cv_score - 2 * std_cv_score #Normal assumption

print(f"Cross-validation Recall Scores: {cv_scores}")
print(f"Cross-validation Mean Recall: {mean_cv_score:.4f}")
print(f"Cross-validation Recall Std Dev: {std_cv_score:.4f}")
print(f"Threshold for Overfitting Detection: {threshold:.4f}")

#Check for overfitting
if test_acc < threshold:
    print("Potential overfitting detected!")
else:
    print("Model generalizes well.")
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import recall_score, classification_report, hinge_loss
import matplotlib.pyplot as plt
import numpy as np

# Parameter
k = 5  # Jumlah fold
epochs = 100
eta = 0.001

# Stratified K-Fold CV
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
recall_scores = []
loss_history_folds = []

for fold, (train_idx, val_idx) in enumerate(skf.split(Xcount_train_transformed_ngr, y_train)):
    print(f"\nFold {fold+1}/{k}")

    X_train_fold = Xcount_train_transformed_ngr[train_idx]
    y_train_fold = y_train[train_idx]
    X_val_fold = Xcount_train_transformed_ngr[val_idx]
    y_val_fold = y_train[val_idx]

    # Inisialisasi ulang model setiap fold
    svm_iter = SGDClassifier(loss='hinge', learning_rate='constant', eta0=eta,
                             max_iter=1, warm_start=True, random_state=42)

    classes = np.unique(y_train)

    loss_history = []

    for epoch in range(epochs):
        svm_iter.partial_fit(X_train_fold, y_train_fold, classes=classes)
        decision_values = svm_iter.decision_function(X_train_fold)
        loss = hinge_loss(y_train_fold, decision_values)
        loss_history.append(loss)

    loss_history_folds.append(loss_history)

    # Evaluasi recall di validation set
    y_val_pred = svm_iter.predict(X_val_fold)
    recall = recall_score(y_val_fold, y_val_pred, average='macro')
    recall_scores.append(recall)

    print(f"Recall (macro average) for fold {fold+1}: {recall:.4f}")
    print(classification_report(y_val_fold, y_val_pred))

# Rata-rata dan standar deviasi recall seluruh fold
mean_recall = np.mean(recall_scores)
std_recall = np.std(recall_scores)

print("\n📊 Hasil Evaluasi Cross-Validation:")
print("Rata-rata Recall (macro): {:.4f}".format(mean_recall))
print("Standar Deviasi Recall   : {:.4f}".format(std_recall))

# Plot loss setiap fold
for i, loss_history in enumerate(loss_history_folds):
    plt.plot(range(1, epochs + 1), loss_history, label=f'Fold {i+1}')

plt.title('Hinge Loss per Epoch untuk Tiap Fold')
plt.xlabel('Epoch')
plt.ylabel('Hinge Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

# Konversi dataset ke format DMatrix (format khusus untuk XGBoost)
dtrain = xgb.DMatrix(X_train_scaled_ngr, label=y_train_enc)
dtest = xgb.DMatrix(X_test_scaled_ngr, label=y_test_enc)

# Hyperparameter
params = {
    'objective': 'multi:softmax',  # Multiclass classification
    'num_class': len(le.classes_),  # Jumlah kelas dalam target
    'eval_metric': 'mlogloss'  # Log loss sebagai metric evaluasi
}

# menyimpan hasil evaluasi
evals_result = {}

# Training model
bst = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=300,  # Jumlah iterasi boosting
    evals=[(dtrain, 'train'), (dtest, 'eval')],
    early_stopping_rounds=10,  # Berhenti jika tidak ada peningkatan
    evals_result=evals_result,  # Menyimpan hasil evaluasi
    verbose_eval=True
)

# Plot loss selama training
plt.plot(evals_result['train']['mlogloss'], label='Training Loss')
plt.plot(evals_result['eval']['mlogloss'], label='Validation Loss')
plt.xlabel("Iterations")
plt.ylabel("Log Loss")
plt.title("Training vs Validation Loss in XGBoost")
plt.legend()
plt.show()
```

```python
# Make predictions on the test set
y_pred = bst.predict(dtest)

# Convert predicted labels back to original class names (if needed)
y_pred_labels = le.inverse_transform(y_pred.astype(int))
y_test_labels = le.inverse_transform(y_test_enc)

# Print classification report
print(classification_report(y_test_labels, y_pred_labels))
print(accuracy_score(y_test_labels, y_pred_labels))
```

```python
import xgboost as xgb
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import recall_score, classification_report

# Encode label
le = LabelEncoder()
y_encoded = le.fit_transform(y_train)

# K-Fold Cross Validation
k = 5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
recall_scores = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_scaled_ngr, y_encoded)):
    print(f"\nFold {fold+1}/{k}")

    # Split data
    X_train_fold = X_train_scaled_ngr[train_idx]
    y_train_fold = y_encoded[train_idx]
    X_val_fold = X_train_scaled_ngr[val_idx]
    y_val_fold = y_encoded[val_idx]

    # Konversi ke DMatrix
    dtrain = xgb.DMatrix(X_train_fold, label=y_train_fold)
    dval = xgb.DMatrix(X_val_fold, label=y_val_fold)

    # Param dan tracking
    params = {
        'objective': 'multi:softmax',
        'num_class': len(le.classes_),
        'eval_metric': 'mlogloss'
    }

    evals_result = {}
    booster = xgb.train(
        params=params,
        dtrain=dtrain,
        num_boost_round=300,
        evals=[(dtrain, 'train'), (dval, 'val')],
        early_stopping_rounds=10,
        evals_result=evals_result,
        verbose_eval=False
    )

    # Predict dan evaluasi
    y_val_pred = booster.predict(dval)
    recall = recall_score(y_val_fold, y_val_pred, average='macro')
    recall_scores.append(recall)

    print(f"Recall (macro average) fold {fold+1}: {recall:.4f}")
    print(classification_report(y_val_fold, y_val_pred, target_names=le.classes_.astype(str)))


    # # Optional: plot loss
    # plt.plot(evals_result['train']['mlogloss'], label=f'Train Fold {fold+1}')
    # plt.plot(evals_result['val']['mlogloss'], label=f'Val Fold {fold+1}')

# Final hasil evaluasi
mean_recall = np.mean(recall_scores)
std_recall = np.std(recall_scores)

# print("\n📊 Hasil Evaluasi XGBoost Cross-Validation:")
print(f"Rata-rata Recall (macro): {mean_recall:.4f}")
print(f"Standar Deviasi Recall   : {std_recall:.4f}")

# # Tampilkan loss curve
# plt.title("Training vs Validation Log Loss (per Fold)")
# plt.xlabel("Boosting Round")
# plt.ylabel("Log Loss")
# plt.legend()
# plt.grid(True)
# plt.show()
```
